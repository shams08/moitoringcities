# SEWER data

```python
# extract sewer complaints
sewer_data = data.loc[data['Complaint Type']=='Sewer']
```

```python
sewer_data.shape
```

```python
#sewer_data.to_csv('sewer_data')
```

```python
# Change column name in a standard format
sewer_data.columns = sewer_data.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# check few rows
sewer_data.head()
```

```python
# Create pandas date time format
sewer_data ["created_date"] = pd.to_datetime (sewer_data.created_date)
```

```python
# extract sewer complaints for 2018
sewer_data_2018 = sewer_data.loc [sewer_data.created_date.dt.year ==2018]
```

```python
sewer_data_2018.shape
```

```python
# Check different kind of sewer complaints. some complaints are related to sewer backup
sewer_data_2018.descriptor.unique()
```

```python
# Create a new dataframe with the sewer backup complaints
complain1 = sewer_data_2018.loc[sewer_data_2018.descriptor== 'Sewer Backup (Use Comments) (SA)']
complain2 = sewer_data_2018.loc[sewer_data_2018.descriptor== 'Catch Basin Clogged/Flooding (Use Comments) (SC)']
complain3 = sewer_data_2018.loc[sewer_data_2018.descriptor== 'Street Flooding (SJ)']
complain4 = sewer_data_2018.loc[sewer_data_2018.descriptor== 'Manhole Overflow (Use Comments) (SA1)']
complain5 = sewer_data_2018.loc[sewer_data_2018.descriptor== 'Sewer Break (SBR)']
complain6 = sewer_data_2018.loc[sewer_data_2018.descriptor== 'Highway Flooding (SH)']
complain7 = sewer_data_2018.loc[sewer_data_2018.descriptor== 'Culvert Blocked/Needs Cleaning (SE)']

sewer_backup_2018 = pd.concat([complain1,complain2, complain3,complain4,complain5,complain6,complain7])
```

```python
sewer_backup_2018.shape
```

```python
sewer_backup_2018.columns
```

```python
# Clean the data
sewer_backup_2018_cleaned = sewer_backup_2018[['created_date', 'complaint_type', 'complaint_type', 'incident_zip', 'latitude', 'longitude']].copy()
```

```python
# save as csv
sewer_backup_2018_cleaned.to_csv("sewer_backup_2018_cleaned.csv")
```

```python
# check few rows of final sewer backup complaint data
sewer_backup_2018_cleaned.head()
```

## SEWER TIME SERIES

```python
# create a new column with date only
sewer_backup_2018_cleaned['date'] = sewer_backup_2018_cleaned['created_date'].dt.date
```

```python
sewer_backup_2018_cleaned.head()
```

```python
# aggregate data based on date
daily_sewer_com = sewer_backup_2018_cleaned.groupby([sewer_backup_2018_cleaned.date]).count()
```

```python
daily_sewer_com.head()
```

```python
#clean data
daily_sewer_com= daily_sewer_com.drop(['complaint_type', 'complaint_type', 'incident_zip', 'longitude', 'date','latitude'], axis=1)
```

```python
#reset index of the dataframe
daily_sewer_com= daily_sewer_com.reset_index()
```

```python
#check few rows of the result
daily_sewer_com.head()
```

```python
daily_sewer_com = daily_sewer_com.rename(index=str, columns={"created_date": "complaints"})
```

```python
# create the chart
import matplotlib.pyplot as plt
plt.figure();
daily_sewer_com.plot(x='date', y='complaints', linewidth=5, fontsize= 22,
                     figsize=(20,8), grid=True, title=('Brooklyn Hourly Pick ups'), x_compat=True)
plt.xlabel('date', fontsize=22)
plt.title('Sewer Backup Complains in NYC 2018',  fontsize=22)
plt.legend(loc=2, prop={'size': 20})
plt.show()
```

```python
#SAVE CSV

daily_sewer_com.to_csv("number of sewer compaints each day.csv")
```
